```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df_can = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.csv")
df_can.head()
```

```python
df_can.set_index('Country', inplace=True)
df_can.index.name = None
```

```python
years = list(map(str, range(1980, 2014)))
years
```

```python
print('Matplotlib version: ', mpl.__version__)
```

```python
print(plt.style.available)
mpl.style.use(['ggplot'])
```

```python
haiti = df_can.loc['Haiti', years]
haiti.head()
```

```python
haiti.plot()
```

```python
haiti.plot(kind='line')

plt.title('Immigration from Haiti')
plt.ylabel('Number of immigrant')
plt.xlabel('Years')

plt.show()
```

```python
haiti.index = haiti.index.map(int)
haiti.plot(kind='line')

plt.title('Immigratrion from Haiti')
plt.ylabel('Number of immigration')
plt.xlabel('Years')

#Syntax: plt.text(x, y, label)
plt.text(2000, 6000, '2010 Earthquake')

plt.show()
```

We can easily add more countries to line plot to make meaningful comparisons immigration from different countries. 

**Question:** Let's compare the number of immigrants from India and China from 1980 to 2013.


Step 1: Get the data set for China and India, and display the dataframe.

```python
df_ci = df_can.loc[['India', 'China'], years]
df_ci
```

Step 2: Plot graph. We will explicitly specify line plot by passing in `kind` parameter to `plot()`.


```python
df_ci.plot(kind='line')
```

```python
df_ci = df_ci.transpose()
df_ci.head()
```

```python
df_ci.index = df_ci.index.map(int)
df_ci.plot(kind='line')

plt.title('Immigration from China and India')
plt.ylabel('Number of immigration')
plt.xlabel('Years')

plt.show()
```

**Question:** Compare the trend of top 5 countries that contributed the most to immigration to Canada.


```python
df_can.sort_values(by='Total', ascending=False, axis=0, inplace=True)
df_top5 = df_can.head()
df_top5
```

```python
df_top5 = df_top5[years].transpose()
df_top5
```

```python
df_top5.index = df_top5.index.map(int) #Let's change the index value of df_top5 to type integer for plotting
df_top5.plot(kind='line', figsize=(14, 8)) #Pass a tuple (x, y) size

plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
```

```python
df_top5.index = df_top5.index.map(int) #Let's change the index value of df_top5 to type integer for plotting
df_top5.plot(kind='area', figsize=(14, 8)) #Pass a tuple (x, y) size

plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
```

```python
count, bin_edges = np.histogram(df_can['2013']) #This function is going to partition the spread of the data in column 2013 in ten bins

df_can['2013'].plot(kind='hist', xticks = bin_edges)

plt.title('Histogram of immigration from 195 countries in 2013')
plt.ylabel('Number of Countries')
plt.xlabel('Number of immigrants')

plt.show()
```
