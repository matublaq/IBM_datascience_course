```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import requests
import seaborn as sns
import folium
```

```python
world_map = folium.Map()
world_map
```

```python
canada_map = folium.Map(
    location=[56.130, -106.35], 
    zoom_start=4, 
    #tiles='Stamen Toner', # or "Stalem Terrain"
    
)

canada_map
```

```python
folium.Marker(location=[51.2538, -85.3232], popup='Ontario').add_to(canada_map)
canada_map
```

```python
#Create a feature group
ontario = folium.map.FeatureGroup()

#style the feature group
ontario.add_child(
    folium.features.CircleMarker([51.25, -85.32], radius=5, color='red', fill_color='red')
    )

#add the feature group to the map
canada_map.add_child(ontario)

canada_map
```

### Show multiple markers points 

```python
#Define a list oof locations and their corresponding popus
locations = [
    {"location": [45.4215, -75.6989], 'popup': 'Ottawa'}, 
    {"location": [53.5461, -113.4938], 'popup': 'Edmonton'}, 
    {"location": [49.2827, -123.1207], 'popup': 'Vancouver'}
]

#Add markers for each location in the list
for loc in locations:
    folium.Marker(location=loc["location"], popup=loc["popup"]).add_to(canada_map)

canada_map
```

MarkCluster object

```python
from folium.plugins import MarkerCluster

#Create a MarkerCluster object
marker_cluster = MarkerCluster().add_to(canada_map)

# Add markers for each location in the list to the MarkerCluster
for loc in locations:
    folium.Marker(location=loc["location"], popup=loc["popup"]).add_to(marker_cluster)
```
